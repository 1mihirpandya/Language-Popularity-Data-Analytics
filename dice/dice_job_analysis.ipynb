```python
def string_in_list(string, substring):
    l = [s.strip().lower() for s in str(string).split(",")]
    return substring in l
```

```python
#JavaScript, SQL Java, C#, PHP, Python, C++, C, nodejs, ruby
```

```python
LANGS_A = ["javascript", "sql", "java", "c#", "php", "python", "c++", "c", "ruby", "node.js", "objective-c"]
```

```python
for lang in LANGS_A:
    dice_df[lang] = dice_df["sector"].apply(lambda string: string_in_list(string, lang))
```

```python
date_df['weekday'] = date_df['date_added'].apply(lambda d: datetime.strptime(d, "%m/%d/%Y").weekday()
)
date_df['month'] = date_df['date_added'].apply(lambda d: datetime.strptime(d, "%m/%d/%Y").month)
date_df
```

```python
month_df = date_df.groupby("month").sum().reset_index().drop(['weekday'], axis=1)
month_df.to_csv('dice_monthly.csv', index=False)
month_df
```

```python
week_df = date_df.groupby("weekday").sum().reset_index().drop(['month'], axis=1)
week_df.to_csv('dice_weekday.csv', index=False)
week_df
```

```python
state_to_tz = {}
tz_shortform = {'Central Time Zone': 'CST',
        'Mountain Time Zone': 'MST',
        'Eastern Time Zone': 'EST',
        'Pacific Time Zone': 'PST'}
#http://www.softschools.com/social_studies/state_abbreviations/
#https://state.1keydata.com/time-zone-state.php
with open('time_zone.txt') as tz_infile:
    for line in tz_infile:
        state = line[:2]
        rest = line[2:]
        first_tz = rest.split(',')[0].strip()
        if first_tz in tz_shortform:
            state_to_tz[state] = tz_shortform[first_tz]
state_to_tz
```

```python
dice_df['state'] = dice_df['location'].apply(lambda l: str(l).split(' ')[-1])
dice_df['timezone'] = dice_df['state'].apply(lambda s: state_to_tz[s] if s in state_to_tz else '')
dice_df
```

```python
tz_df = dice_df[LANGS + ['timezone']][dice_df['timezone'] != '']
tz_df
```

```python
s_timezone_df = tz_df.groupby('timezone').sum().reset_index()
s_timezone_df.to_csv('dice_timezone.csv', index=False)
s_timezone_df
```
